#**Modeling**

```python
from sklearn.preprocessing import FunctionTransformer,StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from imblearn.pipeline import make_pipeline,Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_validate,cross_val_score,cross_val_predict,train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
y=pd.get_dummies(y,drop_first=True)
```

```python
y
```

we used log function to numeric columns & one hot encoder to categorical columns


```python
log_fun=FunctionTransformer(func=np.log1p)
```

```python
transform=ColumnTransformer(transformers=[
    ('log',log_fun,x.select_dtypes('number').columns),
    ('ohe',OneHotEncoder(sparse_output=False,drop='first'),x.select_dtypes('object').columns)
],remainder='passthrough'
)
```

```python
x_transform=transform.fit_transform(x)
```

```python
x_transform
```

```python
y
```

we used pipeline used the transform and Stander Scaler then
see what model will do better

```python
models=[
    ('lr',LogisticRegression()),
    ('SVC',SVC()),
    ('DT',DecisionTreeClassifier()),
    ('RT',RandomForestClassifier())
]
```

```python
for model in models:
  print(model[0])
  PL=make_pipeline(transform,

                  StandardScaler(),
                  model[1])
  cv=cross_validate(estimator=PL,X=x,y=y,cv=7,return_train_score=True,scoring='recall')
  print(model[0],'train score = ',cv['train_score'].mean())
  print(model[0],'test score = ',cv['test_score'].mean())
```

#**Hyperparameter Tuning**

```python
PL=make_pipeline(transform,

                  StandardScaler())
```

```python
x_pre=PL.fit_transform(x)
```

**Random forest Hyper tuning**

```python
params=[
    {'criterion':['gini','entropy'],'n_estimators':range(70,250)}
]

rand_search=RandomizedSearchCV(estimator=RandomForestClassifier(),
                               param_distributions=params,
                               scoring='recall',
                               cv=5)
rand_search.fit(x_pre,y)
```

```python
rand_search.best_score_
```

```python
rand_search.best_params_
```

**Decision tree Hyper tuning**

```python
params=[
    {'max_depth':range(3,15),'max_features':range(3,15)}
]
rand_search=GridSearchCV(estimator=DecisionTreeClassifier(),
                         param_grid=params,
                               scoring='recall',
                               cv=5)
rand_search.fit(x_pre,y)
```

```python
rand_search.best_score_
```

```python
rand_search.best_params_
```

**SVC Hyper tuning**

```python
params=[
    {'C':[1,10,100],'kernel':['linear','sigmoid','poly']},
    {'C':[1,10,100],'kernel':['rbf'],'gamma':[0.5,0.6,0.7,0.1,0.01]}
]
rand_search=GridSearchCV(estimator=SVC(),
                         param_grid=params,
                               scoring='recall',
                               cv=5)
rand_search.fit(x_pre,y)
```

```python
rand_search.best_score_
```

```python
rand_search.best_params_
```

**SO conclusion the SVC is the best model with parameter {'C': 1, 'gamma': 0.7, 'kernel': 'rbf'}**
